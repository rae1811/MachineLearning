Popularity Based Recommendations

```python
ratings_with_info =ratings.merge(books, on ='ISBN')
```

```python
num_rating_df =ratings_with_info.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating' :'num_ratings'}, inplace =True)
num_rating_df
```

```python
avg_rating_df =ratings_with_info.groupby('Book-Title').mean(numeric_only =True)['Book-Rating'].reset_index()
```

```python
avg_rating_df.rename(columns={'Book-Rating' :'avg_ratings'}, inplace =True)
avg_rating_df
```

```python
popular_df =avg_rating_df.merge(num_rating_df, on ='Book-Title')
popular_df
```

```python
popular_df =popular_df[popular_df['num_ratings'] >= 250].sort_values('avg_ratings', ascending=False).head(50)
```

```python
popular_df=popular_df.merge(books, on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M', 'num_ratings','avg_ratings']]
```

```python
popular_df
```

Collaborative Filtering Based Recommender System

```python
x =ratings_with_info.groupby('User-ID').count()['Book-Rating']>200
matrix= x[x].index
```

```python
matrix
```

```python
ratings_with_info['User-ID'].isin(matrix)
```

```python
filtered_ratings = ratings_with_info[ratings_with_info['User-ID'].isin(matrix)]
```

```python
filtered_ratings
```

```python
y=filtered_ratings.groupby('Book-Title').count()['Book-Rating']>=50
famous_books =y[y].index
```

```python
filtered_ratings=filtered_ratings[filtered_ratings['Book-Title'].isin(famous_books)]
```

```python
filtered_ratings
```

```python
pt =filtered_ratings.pivot_table(index='Book-Title', columns='User-ID', values ='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score.shape
```

```python
def recommend(book_name):
    index=np.where(pt.index==book_name)[0][0]
    similar_items =sorted(list(enumerate(similarity_score[index])), key=lambda x:x[1], reverse=True)[1:6]
    data = []

    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    
    return data
```

```python
recommend('1984')
```
